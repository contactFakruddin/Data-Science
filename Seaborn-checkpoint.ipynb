```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.plot(x,y,'r--')
plt.plot(x,y,'k*')
# line style: --,-,-.,:
plt.plot(y,x,'b:')
plt.plot(y,x,'g^')
#line colour: red(r),blue(b),green(g),Yellow(y),Magenta(m),Black(k),Cyan(c)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("Title")
plt.xlim(0,25)
plt.ylim(-5,30)
```

```python
name=table.index
plt.plot(name,table['python'],'r--')
plt.plot(name,table['dsa'],'g--')
plt.xlim(0,5)
plt.ylim(-5,100)
```

```python
#Subplot:

plt.subplot(1,3,1)
plt.plot(name,table['python'],'r--')

plt.subplot(1,3,2)
plt.plot(name,table['dsa'],'c-.')

plt.subplot(1,3,3)
plt.plot(name,table['stb'],'ko-')
```

```python
plt.subplot(2,2,1)
plt.plot(name,table['python'],'r--',label='Python')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("| Python |")
plt.legend(loc=8)

plt.subplot(2,2,2)
plt.plot(name,table['dsa'],'c-.',label='DSA')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("| DSA |")
plt.legend(loc=1)


plt.subplot(2,2,3)
plt.plot(name,table['stb'],'ko-',label='STB')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("| Software Testing |")
plt.legend(loc=5)


plt.subplot(2,2,4)
plt.plot(name,table['ooad'],'m-',label='OOAD')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title("| OOAD |")

plt.tight_layout()
plt.legend(loc=8)
```

```python
#Univariant    #Multivariant
#Pie           #Scatter
#Histogram     #Barchart
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Pie,Scatter,Stairs,Stackplot,Stem,Line,Hist2d,Hexbin:
x=np.random.normal(0,10,200)
y=np.random.normal(0,20,200)
c=np.random.normal(0,20,200)
s=np.random.normal(0,10,200)

plt.scatter(x,y,s=100,c=c,marker='o',alpha=1,cmap='cool',linewidth=1,edgecolor='k')
plt.colorbar()

```

```python
plt.bar(x,y)
```

```python
plt.scatter(x,y,s=100,c=c,marker='p',alpha=1,cmap='cool',linewidth=1,edgecolor='k')
plt.colorbar()
```

```python
plt.hist(x,bins=10)
plt.hist(y,bins=50)
plt.hist(s,bins=70)
plt.hist(c,bins=100)
```

```python
plt.hist2d(x,y,bins=10,cmap='Reds')
```

```python
x=np.random.normal(0,10,25)
y=np.random.normal(0,20,25)
plt.hexbin(x,y,bins=100,cmap='Blues',gridsize=10)
```

```python
plt.stem(x,y)
```

```python
plt.stairs(x)
plt.stairs(s)
```

```python
plt.hist(y,bins=30,histtype='step',alpha=0.5)
plt.hist(x,bins=30,histtype='barstacked',alpha=0.5)
plt.hist(s,bins=30,histtype='bar',alpha=0.5)
plt.hist(c,bins=30,histtype='bar',alpha=0.5)
```

```python
#Rectangular Vertical box-plot: 

d3=np.random.normal(0,200,160)
d1=np.random.normal(0,200,160)
d2=np.random.normal(0,200,160)
d=[d1,d2,d3]

#print(d)
plt.boxplot(d,vert=False,patch_artist=True)
```

```python
#Rectangular Horizontal box-plot:(USING SAME DATA)
#Printing d:
plt.boxplot(d,vert=True,patch_artist=False)
```
